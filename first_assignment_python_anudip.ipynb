4. Write a Python program to calculate the average of four numbers entered by the user.

```python
num1 = int(input("Write First Number out of Four to Calculate the average of four numbers = "))
num2 = int(input("Write second Number out of Four to Calculate the average of four numbers = "))
num3 = int(input("Write Third Number out of Four to Calculate the average of four numbers = "))
num4 = int(input("Write forth Number out of Four to Calculate the average of four numbers = "))

average_of_4_Number = (num1 + num2 + num3 + num4) / 4

print(f"The average of {num1}, {num2}, {num3}, {num4} is = ", average_of_4_Number)
```

5. Write a Python program to swap two numbers using a temporary variable.

```python
num1 = int(input("Write First Number out of Two to swap numbers = "))
num2 = int(input("Write second Number out of Two to swap numbers = "))

temp_var = num1
num1 = num2
num2 = temp_var

print(f"The Swaped Number is {num1}, {num2}")

```
